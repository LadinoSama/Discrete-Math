```
class Automorfismo:

    def __init__(self):
        self.strg = "123456"
        self.permutation = []
        self.auto = []
        self.Permutations(self.strg, 0)
        self.g = self.crearGrafo()
        #print(self.permutation)
        self.createPermGroups(list(self.strg),self.permutation)

    def Permutations(self, string, step):
        # if we've gotten to the end, add the permutation to list permutation
        if step == len(string):
            #print("".join(string))
            self.permutation.append((string))
        # everything to the right of step has not been swapped yet
        for i in range(step, len(string)):
            # copy the string (store as array)
            string_copy = [character for character in string]
            # swap the current index with the step
            string_copy[step], string_copy[i] = string_copy[i], string_copy[step]
            # recurse on the portion of the string that has not been swapped yet (now it's index will begin with step + 1)
            self.Permutations(string_copy, step + 1)

    def crearGrafo(self):
        graph = Grafo()
        for i in range(1,len(self.strg)+1):
            graph.agregarVertice(i)
        graph.agregarArista(1, 2)
        graph.agregarArista(1, 3)
        graph.agregarArista(2, 3)
        graph.agregarArista(2, 1)
        graph.agregarArista(3, 4)
        graph.agregarArista(3, 1)
        graph.agregarArista(3, 2)
        graph.agregarArista(4, 3)
        graph.agregarArista(4, 5)
        graph.agregarArista(4, 6)
        graph.agregarArista(5, 4)
        graph.agregarArista(5, 6)
        graph.agregarArista(6, 4)
        graph.agregarArista(6, 5)
        return graph

    def createPermGroups(self, e, perm):
        auto = []
        for i in perm:
            compDict = {}
            for a in range(len(e)):
                compDict[int(e[a])] = int(i[a])
            l = []
            auxlist = []
            for b in range(len(list(compDict.keys()))):
                aux = self.ordenar(compDict, list(compDict.keys())[b], compDict.get(list(compDict.keys())[b]),l,l)
                if type(aux) != int:
                    ax = self.quitarRepetidos(aux, auxlist)
                    if ax not in auxlist:
                        auxlist.append(ax)
            if auxlist:
                auxlist.pop(0)
            #print(auxlist)
            if self.acept(auxlist) and auxlist not in auto:
                auto.append(auxlist)
        print(self.takeOut(auto))

    def ordenar(self, dict, first, second, lista, auxlist):
        a = first
        b = second
        if b in auxlist:
            return set(auxlist)
        elif a != b:
            if a not in auxlist:
                auxlist.append(a)
            if b not in auxlist:
                auxlist.append(b)
            return self.ordenar(dict, b, dict.get(b), lista, auxlist)
        elif a == b:
            return a

    def acept(self, array):
        acept = True
        for a in array:
            for b in a:
                v = self.g.listaVertices.get(int(b)).conectadoA
                size = len(v)
                for c in a:
                    if len(self.g.listaVertices.get(int(c)).conectadoA) != size:
                        acept = False
            if len(a) % 2 != 0:
                acept = False
        return acept

    def quitarRepetidos(self, con, lista):
        aux = []
        if len(lista) > 0:
            for a in lista:
                aux = con - a
        return set(aux)

    def takeOut(self, lista):
        aux = []
        for a in lista:
            if len(a) > 1:
                aux.append(a)
        for b in lista:
            if len(b) == 1:
                pasa = True
                for c in aux:
                    if b in c:
                        pasa = False
                if pasa == True:
                    aux.append(b)
        aux.insert(0,'e')
        return aux

class Grafo:
    def __init__(self):
        self.listaVertices = {}
        self.numVertices = 0

    def agregarVertice(self, clave):
        self.numVertices = self.numVertices + 1
        nuevoVertice = Vertice(clave)
        self.listaVertices[clave] = nuevoVertice
        return nuevoVertice

    def obtenerVertice(self, n):
        if n in self.listaVertices:
            return self.listaVertices[n]
        else:
            return None

    def __contains__(self, n):
        return n in self.listaVertices

    def agregarArista(self, de, a):
        self.listaVertices.get(de).agregarVecino(a)

    def obtenerVertices(self):
        return self.listaVertices.keys()

    def __iter__(self):
        return iter(self.listaVertices.values())

    def ReturnVerticeInicial(self):
        for i in self.listaVertices:
            if self.listaVertices.get(i).initialstate == True:
                return self.obtenerVertice(i)

class Vertice:
    ##se crea el constructor, self es la instancia del objeto creado de la clase
    def __init__(self,clave):
        self.id = clave
        self.conectadoA = []

    def setinitialstate(self):
        self.initialstate = True

    def setacceptingstate(self):
        self.acceptingstate = True

    def getinitialstate(self):
        return  self.initialstate

    def getacceptingstate(self):
        return  self.acceptingstate

    ##agregar vecino: conecta un vertice a otro
    def agregarVecino(self,vecino):
        self.conectadoA.append(vecino)

    ##el metodo keys arroja todos las llaves del diccionario, donde esas keys son los nombres de los vertices al cual esta conectado el vertice quee stamos llamando
    def obtenerConexiones(self):
        return self.conectadoA.keys()

    def obtenerId(self):
        return self.id


    def obtenerPonderacion(self,vecino):
        return self.conectadoA[vecino]

a = Automorfismo()                                                                                                                                                                
```
